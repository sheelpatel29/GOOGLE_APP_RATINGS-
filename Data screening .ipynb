```python
import pandas as pd 
import numpy as np
import matplotlib.pylab as plt
```

```python
df= pd.read_csv('googleplaystore.csv')
```

```python
df.info()
```

```python
df.isna().sum(axis=0)
```

```python
df['Rating'].isna().sum(axis=0)
```

```python
df.dropna(subset = ['Rating'],how='all',inplace=True)
```

```python
df['Rating'].isna().sum(axis=0)
```

```python
df.head(60)
```

```python
df['Size1']=df.Size.str.replace('k','')
```

```python
df['Size2']=df.Size.str.replace('M','')
```

```python
df
```

```python
df['Size1']=df['Size1'].str.replace('Varies with device','NA')
```

```python
df.head(60)
```

```python
df['Size']=df['Size'].str.replace('Varies with device','NA')
```

```python
df
```

```python
sh= df.Size1.str.extract(r'(?P<val>[0-9.]*)(?P<unit>[M]{0,1})')
```

```python
sh
```

```python
df["Value"]=sh["val"]
```

```python
df["unit"]=sh["unit"]
```

```python
df
```

```python
df.dtypes
```

```python
df["Value"]=pd.to_numeric(df["Value"])
```

```python
unit_map = {'':1, 'M':1e3}
df['Size'] = df.Value.astype(np.float64) * df.unit.map(unit_map)
df
```

```python
df['Installs']=df.Installs.str.split(r"\+|=", expand=True)
```

```python
df["Installs"]
```

```python
df = df.convert_dtypes()
```

```python
df.dtypes
```

```python
df['Installs']=df['Installs'].str.replace('Free','0')
```

```python
df['Installs'] = df['Installs'].apply(lambda x: x.replace(',', '')
                                if isinstance(x, str) else x).astype(float)
```

```python
df['Installs'].unique()
```

```python
df['Price']=df['Price'].str.replace('Everyone','0')
```

```python
df['Price'] = df['Price'].apply(lambda x: x.replace('$', '')
                                if isinstance(x, str) else x).astype(float)
```

```python
df["Price"].unique()
```

```python
df.drop(df[df.Rating>5].index, inplace=True)
```

```python
df.Rating.unique()
```

```python
df.drop(columns=["Size1","Size2","Value","unit"],axis=1,inplace=True)
```

```python
df.info()
```

```python
df["Reviews"]=df.Reviews.astype(np.float64)
```

```python
df=df[df['Reviews'] <= df['Installs']].copy()
```
