```python
import pandas as pd 
import numpy as np
import matplotlib.pylab as plt
```

```python
df=pd.read_csv("App_Ratings_cleaned.csv")
```

```python
#Performing uniVariate analysis 

df.boxplot(column=["Price"],figsize=(12,6))
```

```python
#Note: We can see there are outliers in Price and Reviews, which are to be treated 
```

```python
df[df["Price"]>200]
```

```python
#This seems that apps with greater than 200$ is unwanted and mostly suspicious apps, so drop it 
```

```python
df1=df[df["Price"]<200].copy()
```

```python
df1
```

```python
#Also reviews greater than 2Million are mostly star apps, which is not useful for analysis, so drop them for accuracy
```

```python
df=df1[df1["Reviews"]<2*1e7]
df
```

```python
df=df[df["Installs"]< 5000000.0].copy()
df
```

```python
plt.figure(figsize=(12.8,6),dpi=80)
plt.scatter(df["Rating"],df["Price"])
plt.xlabel("Ratings")
plt.ylabel("Price")
plt.title("Ratings V/S Price Analysis")
```

```python
#Ratings increase with increase in Price 
```

```python
plt.figure(figsize=(12.8,6),dpi=80)
plt.scatter(df["Rating"],df["Size"])
plt.xlabel("Ratings")
plt.ylabel("Size")
plt.title("Ratings V/S Size Analysis")
```

```python
#Larger Size app are rated better 
```

```python
plt.figure(figsize=(12.8,6),dpi=80)
plt.scatter(df["Rating"],df["Reviews"])
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.title("Ratings V/S Reviews Analysis")
```

```python
#It is clearly evident that reviews go higher with ratings in mind 
```
